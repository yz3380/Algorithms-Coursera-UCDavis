```python
# Problem1 Money Change

def get_change(m):
    n = m // 10
    m = m % 10
    n = n + m // 5
    m = m % 5
    n = n + m
    return n

if __name__ == '__main__':
    m = int(input())
    print(get_change(m))
```

```python
# Problem2 Max Value of the Loot

def get_optimal_value(capacity, weights, values):
    value = 0.
    n = len(weights)
    value_per_weight=[]
    for i in range(n):
        value_per_weight.append((values[i] / weights[i], values[i], weights[i]))
    
    value_per_weight.sort(reverse=True)
    
    for i in range(n):
        if capacity == 0:
            return value
        a = min(value_per_weight[i][2], capacity)
        value = value + a * value_per_weight[i][0]
        capacity = capacity - a

    return value


if __name__ == "__main__":
    data = list(map(int, input().split()))
    n, capacity = data[0:2]
    values = data[2:(2 * n + 2):2]
    weights = data[3:(2 * n + 2):2]
    opt_value = get_optimal_value(capacity, weights, values)
    print("{:.10f}".format(opt_value))
```

```python
# Problem3 Car Fueling

def compute_min_refills(distance, tank, stops):

    n = len(stops)
    numRefills = 0
    currentRefill = 0
    stops.insert(0,0) # insert 0 at the start of stops
    stops.append(distance) # add distance at the end
    #print(stops)
    while currentRefill <= n:
        lastRefill = currentRefill
        while currentRefill <= n and (stops[currentRefill+1] - stops[lastRefill]) <= tank:
            currentRefill = currentRefill + 1
            #print(currentRefill)
        if currentRefill == lastRefill:
            return -1
        if currentRefill <= n:
            numRefills = numRefills + 1
    return numRefills

if __name__ == '__main__':
    d, m, n, *stops = map(int, input().split())
    print(compute_min_refills(d, m, stops))

```

```python
# Problem4 Maximum Advertisement Revenue

def max_dot_product(a, b):
    
    res = 0
    a.sort()
    b.sort()
    for i in range(len(a)):
        res += a[i] * b[i]
    return res

if __name__ == '__main__':
    
    data = list(map(int, input().split()))
    n = data[0]
    a = data[1:(n + 1)]
    b = data[(n + 1):]
    print(max_dot_product(a, b))
```

```python
# Problem5 Collecting Signatures

import sys
from collections import namedtuple

Segment = namedtuple('Segment', 'start end')

def optimal_points(segments):
    points = []
    segments.sort(key = lambda x:x[1]) # sort by end points
    #print(segments)
    s = 0
    while s < len(segments): # loop through all segments
        end = segments[s][1]
        points.append(end)
        s = s + 1
        while s < len(segments) and segments[s][0] <= end:
            s = s + 1
       
    return points

if __name__ == '__main__':
    #input = sys.stdin.read()
    n, *data = map(int, input().split())
    segments = list(map(lambda x: Segment(x[0], x[1]), zip(data[::2], data[1::2])))
    points = optimal_points(segments)
    print(len(points))
    for p in points:
        print(p, end=' ')
```

```python
# Problem6 Maximum Number of Prizes

import sys

def optimal_summands(n):
    summands = []
    i = 0
    total = 0
    while total < n:
        i = i + 1
        last = total
        total = total + i
        if (n - total) < (i + 1):
            summands.append(n - last)
            break
        else:
            summands.append(i)
       
    return summands

if __name__ == '__main__':
    #input = sys.stdin.read()
    n = int(input())
    summands = optimal_summands(n)
    print(len(summands))
    for x in summands:
        print(x, end=' ')
```

```python
# Problem7 Maximum Salary

import sys

def IsGreaterOrEqual(digit, maxDigit):
    if maxDigit == -1:
        return True
    digit = str(digit)
    maxDigit = str(maxDigit)
    x = digit + maxDigit
    y = maxDigit + digit
    return(int(x) >= int(y))
    
def largest_number(a):
    res = ""
    while len(a) > 0:
        maxDigit = -1
        maxRank = -1
        for i in range(len(a)):
            if IsGreaterOrEqual(a[i], maxDigit):
                maxDigit = a[i]
                maxRank = i
        res += maxDigit
        a.pop(maxRank)
    
    return int(res)

if __name__ == '__main__':
    #input = sys.stdin.read()
    data = input().split()
    a = data[1:]
    print(largest_number(a))
```
