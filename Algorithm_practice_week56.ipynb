```python
# Problem1 Money Change Again

import sys

def get_change(m, Coins):
    n = len(Coins)
    MinNumCoins = []
    MinNumCoins.append(0)
    
    for i in range(1, m + 1):
        MinNumCoins.append(float('Inf'))
        for j in range(n):
            if i >= Coins[j]:
                NumCoins = MinNumCoins[i - Coins[j]] + 1
                if NumCoins < MinNumCoins[i]:
                    MinNumCoins[i] = NumCoins
    
    return MinNumCoins[m]

if __name__ == '__main__':
    m = int(input())
    Coins = [1 ,3 ,4]
    print(get_change(m, Coins))
```

```python
# Problem2 Primitive Calculator

import sys

def optimal_sequence(n):
    
    MinSteps = []
    MinSteps.append(0)
    
    for i in range(1, n + 1):
        MinSteps.append(float('Inf'))
        if i % 3 == 0:
            NumSteps = MinSteps[i // 3] + 1
            if NumSteps < MinSteps[i]:
                MinSteps[i] = NumSteps
        if i % 2 == 0:
            NumSteps = MinSteps[i // 2] + 1
            if NumSteps < MinSteps[i]:
                MinSteps[i] = NumSteps
        if i >= 1:
            NumSteps = MinSteps[i - 1] + 1
            if NumSteps < MinSteps[i]:
                MinSteps[i] = NumSteps
    
    sequence = output(n, MinSteps)
    
    return reversed(sequence)

def output(n, MinSteps):
    sequence = []
    sequence.append(n)
    while n >= 1:
        if n % 3 == 0 and MinSteps[n // 3] == MinSteps[n] - 1:
            sequence.append(n // 3)
            n = n // 3
        elif n % 2 == 0 and MinSteps[n // 2] == MinSteps[n] - 1:
            sequence.append(n // 2)
            n = n // 2
        else:
            sequence.append(n - 1)
            n = n - 1
    sequence.pop(len(sequence) - 1)
    return sequence
    
#input = sys.stdin.read()
n = int(input())
sequence = list(optimal_sequence(n))
print(len(sequence) - 1)
for x in sequence:
    print(x, end=' ')
```

```python
# Problem3 Edit Distance

def edit_distance(s, t):
    m = len(t)
    n = len(s)
    D = {}
    D[(0,0)] = 0
    for k in range(1, m + 1):
        D[(0, k)] = k
    for k in range(1, n + 1):
        D[(k, 0)] = k
        
    for j in range(1, m + 1):
        for i in range(1, n + 1):
            insertion = D[(i, j - 1)] + 1
            deletion =  D[(i - 1, j)] + 1
            match = D[(i - 1, j - 1)]
            mismatch = D[(i - 1, j - 1)] + 1
            if s[i - 1] == t[j - 1]:
                D[(i, j)] = min(insertion, deletion, match)
            else:
                D[(i, j)] = min(insertion, deletion, mismatch)
    return D[(n, m)]

if __name__ == "__main__":
    print(edit_distance(input(), input()))
```

```python
# Problem4 Longest Common Subsequence of Two

import sys

def lcs2(s, t):
    m = len(t)
    n = len(s)
    D = {}
    D[(0, 0)] = 0
    for k in range(1, m + 1):
        D[(0, k)] = 0
    for k in range(1, n + 1):
        D[(k, 0)] = 0
        
    for j in range(1, m + 1):
        for i in range(1, n + 1):
            insertion =  D[(i, j - 1)]
            deletion =  D[(i - 1, j)]
            match = D[(i - 1, j - 1)] + 1
            mismatch = D[(i - 1, j - 1)]
            if s[i - 1] == t[j - 1]:
                D[(i, j)] = max(insertion, deletion, match)
            else:
                D[(i, j)] = max(insertion, deletion, mismatch)
        
    return D[(n, m)]
    
if __name__ == '__main__':
    #input = sys.stdin.read()
    data = list(map(int, input().split()))

    n = data[0]
    data = data[1:]
    a = data[:n]

    data = data[n:]
    m = data[0]
    data = data[1:]
    b = data[:m]
    
    print(lcs2(a, b))
```

```python
# Problem5 Longest Common Subsequence of Three

import sys

def lcs3(a, b, c):
    l = len(a)
    m = len(b)
    n = len(c)
    D = {}
    
    for i in range(0, l + 1):
        for j in range(0, m + 1):
            D[(i, j, 0)] = 0
            
    for i in range(0, l + 1):
        for k in range(0, n + 1):
            D[(i, 0, k)] = 0
            
    for j in range(0, m + 1):
        for k in range(0, n + 1):
            D[(0, j, k)] = 0
        
    for i in range(1, l + 1):
        for j in range(1, m + 1):
            for k in range(1, n + 1):
                A = D[(i - 1, j, k)]
                B = D[(i, j - 1, k)]
                C = D[(i, j, k - 1)]
                match = D[(i - 1, j - 1, k - 1)] + 1
                mismatch = D[(i - 1, j - 1, k - 1)]
                if a[i - 1] == b[j - 1] == c[k - 1]:
                    D[(i, j, k)] = max(A, B, C, match)
                else:
                    D[(i, j, k)] = max(A, B, C, mismatch)
        
    return D[(l, m, n)]

if __name__ == '__main__':
    #input = sys.stdin.read()
    data = list(map(int, input().split()))
    an = data[0]
    data = data[1:]
    a = data[:an]
    data = data[an:]
    bn = data[0]
    data = data[1:]
    b = data[:bn]
    data = data[bn:]
    cn = data[0]
    data = data[1:]
    c = data[:cn]
    print(lcs3(a, b, c))
```

```python
# Problem1 Maximum Amount of Gold

import sys

def optimal_weight(W, w):
    n = len(w)
    value = {}
    for k in range(W + 1):
        value[(k, 0)] = 0 # only store current and previous column
        value[(k, 1)] = 0       
    
    for j in range(1, n + 1):
        for i in range(1, W + 1):
            value[(i, 0)] = value[(i, 1)] # previous = current
        for i in range(1, W + 1):
            if w[j - 1] <= i:
                value[(i, 1)] = max(value[(i, 1)], value[(i - w[j - 1], 0)] + w[j - 1])  # update new current
    
    return value[(W, 1)]

if __name__ == '__main__':
    #input = sys.stdin.read()
    W, n, *w = list(map(int, input().split()))
    print(optimal_weight(W, w))
```

```python
# Problem2 Partitioning Souvenirs (3 sum)

import sys
#import itertools

def partition3(A):
    sums = sum(A)
    if sums % 3 != 0:
        return 0
    else:
        W = sums // 3
        n = len(A)

    value = {}
    
    # initialize 
    for i in range(W + 1):
        value[(i, 0, 0)] = 0
        value[(0, i, 0)] = 0
        for j in range(W + 1):
            value[(i, j, 0)] = 0
    for k in range(n + 1):
        value[(0, 0, k)] = 1
    
    value[(0, 0, 0)] = 1
    
    for i in range(1, W + 1):
        for k in range(1, n + 1):
            
            # partition 1 initialization
            value[(i, 0, k)] = value[(i, 0, k - 1)]
            if A[k - 1] <= i:
                value[(i, 0, k)] = max(value[(i, 0, k)], value[(i - A[k - 1], 0, k - 1)])
            
            # partition 2 initialization
            value[(0, i, k)] = value[(0, i, k - 1)]
            if A[k - 1] <= i:
                value[(0, i, k)] = max(value[(0, i, k)], value[(0, i - A[k - 1], k - 1)])
    
    # start iteration
    for i in range(1, W + 1):
        for j in range(1, W + 1):
            for k in range(1, n + 1):
                value[(i, j, k)] = value[(i, j, k - 1)]
                if A[k - 1] <= i:
                    value[(i, j, k)] = max(value[(i, j, k)], value[(i - A[k - 1], j, k - 1)])
                if A[k - 1] <= j:
                    value[(i, j, k)] = max(value[(i, j, k)], value[(i, j - A[k - 1], k - 1)])
    
    return value[(W, W, n)]


if __name__ == '__main__':
    #input = sys.stdin.read()
    n, *A = list(map(int, input().split()))
    print(int(partition3(A)))
```

```python
# Problem3 Maximum Value of an Arithmetic Expression

def evalt(a, b, op):
    a = int(a)
    b = int(b)
    if op == '+':
        return a + b
    elif op == '-':
        return a - b
    elif op == '*':
        return a * b
    else:
        assert False

def get_maximum_value(dataset): 
    
    n = len(dataset) // 2 + 1
    if n == 2:
        return evalt(dataset[0:1], dataset[2:3], dataset[1])
    
    m = {}
    M = {}
    
    for i in range(1, n + 1):
        m[(i,i)] = dataset[2 * i - 2]
        M[(i,i)] = dataset[2 * i - 2]

    for s in range(1, n):
        for i in range(1, n + 1 - s):
            j = i + s
            Min = float('Inf')
            Max = -float('Inf')
            for k in range(i, j): 
                op = dataset[2 * k - 1]
                a = evalt(M[(i, k)], M[(k + 1, j)], op)
                b = evalt(M[(i, k)], m[(k + 1, j)], op)
                c = evalt(m[(i, k)], M[(k + 1, j)], op)
                d = evalt(m[(i, k)], m[(k + 1, j)], op)
                Min = min(Min, a, b, c, d)
                Max = max(Max, a, b, c, d) 
                M[(i, j)] = Max
                m[(i, j)] = Min
    
    return M[(1, n)]


if __name__ == "__main__":
    print(get_maximum_value(input()))
                    
```
