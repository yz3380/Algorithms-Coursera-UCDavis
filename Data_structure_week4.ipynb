```python
# Week4
# Problem1 Phone book

class Query:
    def __init__(self, query):
        self.type = query[0]
        self.number = int(query[1])
        if self.type == 'add':
            self.name = query[2]
        

def read_queries():
    n = int(input())
    return [Query(input().split()) for i in range(n)]

def write_responses(result):
    print('\n'.join(result))

def process_queries(queries):
    result = []
    #names = set()
    # Keep list of all existing (i.e. not deleted yet) contacts.
    contacts = {}
    for cur_query in queries:
        if cur_query.type == 'add':
            # if we already have contact with such number,
            # we should rewrite contact's name
            contacts[cur_query.number] = cur_query.name

        elif cur_query.type == 'del':
            if cur_query.number in contacts:
                del contacts[cur_query.number]
                
        else:
            response = 'not found'
            if cur_query.number in contacts:
                response = contacts[cur_query.number]
            result.append(response)
    
    return result

if __name__ == '__main__':
    write_responses(process_queries(read_queries()))
```

```python
# Problem2 Hashing with chains

class Query:

    def __init__(self, query):
        self.type = query[0]
        if self.type == 'check':
            self.ind = int(query[1])
        else:
            self.s = query[1]


class QueryProcessor:
    _multiplier = 263
    _prime = 1000000007

    def __init__(self, bucket_count):
        self.bucket_count = bucket_count # cardinality
        # store all strings in one list
        self.elems = {}
        for i in range(self.bucket_count):
            self.elems[i] = []
            

    def _hash_func(self, s):
        ans = 0
        for c in reversed(s):
            ans = (ans * self._multiplier + ord(c)) % self._prime
        return ans % self.bucket_count

    def write_search_result(self, was_found):
        print('yes' if was_found else 'no')

    def write_chain(self, chain):
        print(' '.join(chain))

    def read_query(self):
        return Query(input().split())

    def process_query(self, query):
        if query.type == "check":
            # use reverse order, because we append strings to the end
            self.write_chain(word for word in reversed(self.elems[query.ind]))
        
        else:
            ind = self._hash_func(query.s)
            if query.type == 'find':
                self.write_search_result(query.s in self.elems[ind])
            elif query.type == 'add':
                if not query.s in self.elems[ind]:
                    self.elems[ind].append(query.s)
            else:
                if query.s in self.elems[ind]:
                    self.elems[ind].remove(query.s)

    def process_queries(self):
        n = int(input())
        for i in range(n):
            self.process_query(self.read_query())

if __name__ == '__main__':
    bucket_count = int(input())
    proc = QueryProcessor(bucket_count)
    proc.process_queries()
```

```python
# Problem3 Find pattern in text

def read_input():
    return (input().rstrip(), input().rstrip())

def print_occurrences(output):
    print(' '.join(map(str, output)))

def PolyHash(S, p, x):
    _hash = 0
    for c in reversed(S):
        _hash = (_hash * x + ord(c)) % p
    return _hash

def PreComputeHashes(T, P, p, x):
    n = len(T)
    H = []
    S = T[(n-P):]
    H.append(PolyHash(S, p, x))
    y = 1
    for i in range(P):
        y = (y * x) % p
    for i in reversed(range(n - P)):
        value = (x * H[-1] + ord(T[i]) - y * ord(T[i + P])) % p
        H.append(value)
    
    H.reverse()
    
    return H
    
    
def get_occurrences(pattern, text):
    _multiplier = 263
    _prime = 1000000007
    P = len(pattern)
    result = []
    pHash = PolyHash(pattern, _prime, _multiplier)
    H = PreComputeHashes(text, P, _prime, _multiplier)
    for i in range(len(H)):
        if pHash != H[i]:
            continue
        if pattern == text[i:(i + P)]:
            result.append(i)
    
    return result
        
    
if __name__ == '__main__':
    print_occurrences(get_occurrences(*read_input()))
```

```python
# Problem5 Substring equality

#import sys
import random

class Solver:
    def __init__(self, s):
        self.s = s
        self.m1 = 10 ** 9 + 7
        self.m2 = 10 ** 9 + 9
        self.x = random.randint(1, 10 ** 9)
        self.H1 = []
        self.H1.append(0)
        self.H2 = []
        self.H2.append(0)
        self.mm1 = []
        self.mm1.append(1)
        self.mm2 = []
        self.mm2.append(1)
        
    def ask(self, a, b, l):
        if a == b:
            return True
        
        n = len(self.H1)
        c = max(a, b)
        
        if n < c + l + 1:
            for i in range(n, c + l + 1):
                self.H1.append((self.x * self.H1[-1] + ord(self.s[i - 1])) % self.m1)
                self.H2.append((self.x * self.H2[-1] + ord(self.s[i - 1])) % self.m2)
        
        n2 = len(self.mm1)
        if n2 < l + 1:
            for i in range(n2, l + 1):
                self.mm1.append((self.x * self.mm1[-1]) % self.m1)
                self.mm2.append((self.x * self.mm2[-1]) % self.m2)
            
        Ha1 = (self.H1[a + l] - self.H1[a] * self.mm1[l]) % self.m1
        Hb1 = (self.H1[b + l] - self.H1[b] * self.mm1[l]) % self.m1
        Ha2 = (self.H2[a + l] - self.H2[a] * self.mm2[l]) % self.m2
        Hb2 = (self.H2[b + l] - self.H2[b] * self.mm2[l]) % self.m2

        if Ha1 == Hb1 and Ha2 == Hb2:
            return True
        
        return False


s = input()
q = int(input())
solver = Solver(s)
for i in range(q):
    a, b, l = map(int, input().split())
    print("Yes" if solver.ask(a, b, l) else "No")
```
